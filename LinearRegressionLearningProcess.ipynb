```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Ground truth parameters
w_true = 2
b_true = 3
```

```python
# Generate 20 random x values
np.random.seed(0)  # For reproducibility
x_values = np.random.uniform(0, 10, 20)
```

```python
# Calculate corresponding y values with noise
y_values       = w_true * x_values + b_true
noise          = np.random.uniform(-0.1, 0.1, y_values.shape) * y_values
y_values_noisy = y_values + noise
```

```python
# Plot the ground truth line and the noisy data points
plt.figure(figsize=(10, 6))
x_line = np.linspace(0, 10, 100)
y_line = w_true * x_line + b_true
```

```python
plt.plot(x_line, y_line, color='green', label='Ground Truth Line')
plt.scatter(x_values, y_values_noisy, color='black', label='Noisy Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Implement Linear Regression with Gradient Descent 
w = np.random.rand()
b = np.random.rand()
```

```python
# Learning Curve 
eta = 0.000001
```

```python
# Initialize the weight and bias
np.random.seed(0)
w = np.random.rand()
b = np.random.rand()
```

```python
# Learning rate
eta = 0.000001
```

```python
# Training data
X = x_values
Y = y_values_noisy
m = len(X)
```

```python
# Training loop
epochs      = 500
mse_history = []
```

```python
for epoch in range(epochs):
    y_pred = w * X + b
    error  = y_pred - Y
    err_b  = np.mean(error)
    err_w  = np.mean(error * X)
    
    # Update weight and bias
    b -= eta * err_b
    w -= eta * err_w
    
    # Calculate mean square error
    mse = np.mean(error**2)
    mse_history.append(mse)
    
    # Visualization update (not animated)
    if epoch % 20 == 0:
        # animation code TODO:
        
        plt.figure(figsize=(10, 6))
        plt.plot(x_line, w * x_line + b, color='orange', label=f'Fitted Line at Epoch {epoch}')
        plt.scatter(X, Y, color='black', label='Noisy Data Points')
        plt.plot(x_line, y_line, color='green', label='Ground Truth Line')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.title(f'Epoch {epoch}, MSE: {mse:.4f}')
        plt.show()
```
